```python
# Selecione apenas as linhas dos paises da oceania com area acima de 200 mil Km2
v = (flags['landmass'] == 6) & (flags['area'] > 200)
flags_oceania_200 = flags[v]

# projeta apenas as colunas "name", "colours", "language", "landmass", "area"
flags_oceania_200 = flags_oceania_200[["name", "colours", "language", "landmass", "area"]]
flags_oceania_200

# Resera o indice
flags_oceania_200 = flags_oceania_200.reset_index(drop=True)
flags_oceania_200
```
