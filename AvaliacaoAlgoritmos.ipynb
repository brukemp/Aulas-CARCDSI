```python
import pickle
from sklearn.model_selection import GridSearchCV, cross_val_score, KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')

caminho = "/content/drive/MyDrive/ADS/DataScience/Source/"
```

```python
with open(caminho+'credit.pkl', 'rb') as f:
  X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
X_credit = np.concatenate((X_credit_treinamento, X_credit_teste), axis = 0)
y_credit = np.concatenate((y_credit_treinamento, y_credit_teste), axis = 0)

X_credit.shape, y_credit.shape
```

```python
## Arvore de decisão


parametros = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 5, 10]
}
```

```python
gri_search = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=parametros,
)

gri_search.fit(X_credit, y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_

melhores_parametros, melhor_resultado
```

```python
# random forest

parametros = {
    'criterion': ['gini', 'entropy'],
    'n_estimators': [10, 40, 100, 150],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 5, 10]
}
```

```python
gri_search = GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid=parametros,
)

gri_search.fit(X_credit, y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_

melhores_parametros, melhor_resultado
```

```python
# KNN

parametros = {
    'n_neighbors': [3, 5,10, 20],
    'p': [1, 2]
}
```

```python
gri_search = GridSearchCV(
    estimator=KNeighborsClassifier(),
    param_grid=parametros,
)

gri_search.fit(X_credit, y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_

melhores_parametros, melhor_resultado
```

```python
# Regressão Logistica

parametros = {
    'tol': [0.0001, 0.00001, 0.000001],
    'C': [1.0, 1.5, 2.0],
    'solver': ['lbfgs', 'sag', 'saga']
}

gri_search = GridSearchCV(
    estimator=LogisticRegression(),
    param_grid=parametros,
)

gri_search.fit(X_credit, y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_

melhores_parametros, melhor_resultado
```

```python
# SVM

parametros = {
    'tol': [0.001, 0.0001, 0.00001],
    'C': [1.0, 1.5, 2.0],
    'kernel': ['rbf', 'linear', 'poly', 'sigmod']
}

gri_search = GridSearchCV(
    estimator=SVC(),
    param_grid=parametros,
)

gri_search.fit(X_credit, y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_

melhores_parametros, melhor_resultado
```

```python
# Redes Neurais

parametros = {
    'activation': ['relu', 'logistic', 'tahn'],
    'solver': ['adam', 'sgd'],
    'batch_size': [10, 56]
}


gri_search = GridSearchCV(
    estimator=MLPClassifier(),
    param_grid=parametros,
)

gri_search.fit(X_credit, y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_

melhores_parametros, melhor_resultado
```

```python
# Validação Cruzada

resultados_arvore = []
resulados_random_forest = []
rresultados_knn = []
resultados_regressao_logistica = []
resultados_svm = []
resultados_redes_neurais = []

for i in range(30):
  print(i)
  kfold = KFold(n_splits=10, shuffle=True, random_state=i)

  arvore = DecisionTreeClassifier(criterion='entropy', min_samples_leaf=1, min_samples_split=5, splitter='best')
  scores = cross_val_score(arvore, X_credit, y_credit, cv=kfold)
  resultados_arvore.append(scores.mean())

  random_forest = RandomForestClassifier(criterion='entropy', min_samples_leaf=1, min_samples_split=5, n_estimators=100)
  scores = cross_val_score(random_forest, X_credit, y_credit, cv=kfold)
  resulados_random_forest.append(scores.mean())

  knn = KNeighborsClassifier(n_neighbors=5, p=2)
  scores = cross_val_score(knn, X_credit, y_credit, cv=kfold)
  rresultados_knn.append(scores.mean())

  regressao_logistica = LogisticRegression(C=2.0, solver='lbfgs', tol=0.0001)
  scores = cross_val_score(regressao_logistica, X_credit, y_credit, cv=kfold)
  resultados_regressao_logistica.append(scores.mean())

  svm = SVC(C=2.0, kernel='rbf', tol=0.001)
  scores = cross_val_score(svm, X_credit, y_credit, cv=kfold)
  resultados_svm.append(scores.mean())

  redes_neurais = MLPClassifier(activation='relu', batch_size=10, solver='adam')
  scores = cross_val_score(redes_neurais, X_credit, y_credit, cv=kfold)
  resultados_redes_neurais.append(scores.mean())

```
