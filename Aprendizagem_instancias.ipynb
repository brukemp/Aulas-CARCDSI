```python
import pickle
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
from sklearn.neighbors import KNeighborsClassifier
```

```python
from google.colab import drive
drive.mount('/content/drive')

caminho = "/content/drive/MyDrive/MaterialApoio/Source/"
```

```python
with open(caminho + 'credit.pkl', 'rb') as f:
  x_credit_treino, y_credit_treino, x_credit_teste, y_credit_teste = pickle.load(f)
```

```python
knn_credit = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn_credit.fit(x_credit_treino, y_credit_treino)
```

```python
previsao_credit = knn_credit.predict(x_credit_teste)
previsao_credit
```

```python
cm = ConfusionMatrix(knn_credit)
cm.fit(x_credit_treino, y_credit_treino)
cm.score(x_credit_teste, y_credit_teste)
```

```python
with open(caminho + 'census.pkl', 'rb') as f:
  x_census_treino, y_census_treino, x_census_teste, y_census_teste = pickle.load(f)
```

```python
knn_census = KNeighborsClassifier(n_neighbors=10)
knn_census.fit(x_census_treino, y_census_treino)
```

```python
previsao_census = knn_census.predict(x_census_teste)
previsao_census
```

```python
accuracy_score(y_census_teste, previsao_census)
```

```python
cm_census = ConfusionMatrix(knn_census)
cm_census.fit(x_census_treino, y_census_treino)
cm_census.score(x_census_teste, y_census_teste)
```
